## 0. Imports y configuración

```python
import pandas as pd
import numpy as np
import sys
sys.path.insert(0, '..')
from utils import load_and_clean, compute_derived_metrics, get_outputs_path

print('Librerías cargadas correctamente')
```

## 1. Carga del dataset raw

```python
# Cargar CSV original sin modificar
df_raw = pd.read_csv('../data/raw/dsb_ictiu01_eurostat_2024.csv')
print(f'Shape: {df_raw.shape}')
df_raw.head(3)
```

## 2. Inspección inicial

```python
# Tipos de datos y nulos
print('Tipos de datos:')
print(df_raw.dtypes)
print(f'\nNulos por columna:')
print(df_raw.isnull().sum())
```

## 3. Limpieza y renombrado de columnas

```python
# Usar función de helpers.py
df = load_and_clean('../data/raw/dsb_ictiu01_eurostat_2024.csv')
print(f'Dataset limpio: {df.shape}')
print(f'Columnas: {df.columns.tolist()}')
df.head()
```

## 4. Gestión de flags y valores faltantes

```python
# Inventario de problemas
print('=== Observaciones con flag u ===')
print(df[df['flag_calidad']=='u'][['pais_cod','categoria_cod','pct_uso_internet','flag_calidad']].to_string())
print(f'\nTotal fiables: {df["fiable"].sum()}')
print(f'Total baja fiabilidad/nulos: {(~df["fiable"]).sum()}')
```

## 5. Exportar dataset limpio

```python
# Guardar processed/dsb_clean.csv
df.to_csv('../data/processed/dsb_clean.csv', index=False)
print('✓ data/processed/dsb_clean.csv guardado')
```

## 6. Calcular y exportar métricas derivadas

```python
# Calcular métricas por país
df_metrics = compute_derived_metrics(df)
print(df_metrics[['pais_cod','gap_total','gap_genero','gap_edad','pos_relativa_eu']].to_string())

# Guardar
df_metrics.to_csv('../data/processed/dsb_derived_metrics.csv', index=False)
print('\n✓ data/processed/dsb_derived_metrics.csv guardado')
```

## 7. Resumen de la limpieza

```python
print('=== RESUMEN PIPELINE ===')
print(f'Filas originales: {df_raw.shape[0]}')
print(f'Columnas originales: {df_raw.shape[1]}')
print(f'Columnas tras limpieza: {df.shape[1]}')
print(f'Observaciones válidas: {df["fiable"].sum()}')
print(f'Observaciones excluidas (nulos): {df["pct_uso_internet"].isna().sum()}')
print(f'Observaciones con baja fiabilidad (flag u): {(df["flag_calidad"]=="u").sum()}')
```

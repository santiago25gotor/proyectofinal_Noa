## 0. Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.insert(0, '..')
from utils import setup_style, PALETTE, save_figure
setup_style()

df = pd.read_csv('../data/processed/dsb_clean.csv')
df_m = pd.read_csv('../data/processed/dsb_derived_metrics.csv')
```

## 1. Brecha de género dentro de DIS_SEV — todos los países

```python
data_gen = df_m[df_m.pais_cod != 'EU27_2020'][['pais_nombre_es','f_dis_sev','m_dis_sev','gap_genero']].sort_values('gap_genero', ascending=False)
print(data_gen.to_string(index=False))
```

## 2. Doble vulnerabilidad en España — gráfico

```python
es = df[(df.pais_cod=='ES') & (df.grupo_edad=='Total') & (df.sexo.isin(['Total','Femenino','Masculino']))
        & (df.dis_nivel.isin(['Sin discapacidad','Limitación severa']))]

fig, ax = plt.subplots(figsize=(8, 5))
pivot_es = es.pivot_table(index='dis_nivel', columns='sexo', values='pct_uso_internet')
pivot_es.plot(kind='bar', ax=ax, color=[PALETTE['blue'],PALETTE['red'],PALETTE['green']], edgecolor='white')
ax.set_title('Doble vulnerabilidad: género × discapacidad\n(España, 2024)', fontweight='bold', color=PALETTE['navy'])
ax.set_ylabel('% que usa Internet')
ax.set_ylim(60, 105)
plt.xticks(rotation=0)
plt.tight_layout()
save_figure(fig, 'fig_doble_vulnerabilidad_espana.png')
plt.show()
```

## 3. Análisis por grupo de edad × discapacidad (España)

```python
es_edad = df[(df.pais_cod=='ES') & (df.sexo=='Total') & (df.grupo_edad.isin(['25-54','55-74']))
             & (df.dis_nivel.isin(['Sin discapacidad','Limitación leve','Limitación severa']))]

print(es_edad[['grupo_edad','dis_nivel','pct_uso_internet']].sort_values(['grupo_edad','dis_nivel']).to_string(index=False))
```

## 0. Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.insert(0, '..')
from utils import setup_style, PALETTE, COUNTRY_NAMES_ES, save_figure
setup_style()

df = pd.read_csv('../data/processed/dsb_clean.csv')
print(f'Dataset cargado: {df.shape}')
```

## 1. Estadísticos descriptivos globales

```python
print(df['pct_uso_internet'].describe().round(2))
```

## 2. Estadísticos por nivel de discapacidad

```python
stats = df[df['sexo']=='Total'][df['grupo_edad']=='Total'].groupby('dis_nivel')['pct_uso_internet'].describe().round(2)
print(stats)
```

## 3. Distribución de valores — boxplot por nivel de discapacidad

```python
fig, ax = plt.subplots(figsize=(10, 5))
orden = ['Sin discapacidad','Limitación leve','Limitación leve o severa (agrupada)','Limitación severa']
df_box = df[df['dis_nivel'].isin(orden) & (df['sexo']=='Total') & (df['grupo_edad']=='Total')]
sns.boxplot(data=df_box, x='dis_nivel', y='pct_uso_internet', order=orden,
            palette=[PALETTE['green'],PALETTE['blue'],PALETTE['amber'],PALETTE['red']], ax=ax)
ax.set_title('Distribución del uso de Internet por nivel de discapacidad\n(todos los países, 2024)', fontweight='bold', color=PALETTE['navy'])
ax.set_xlabel('')
ax.set_ylabel('% de individuos que usan Internet')
plt.xticks(rotation=10)
fig.text(0.01, -0.02, 'Fuente: elaboración propia a partir de Eurostat DSB_ICTIU01 (2024).', fontsize=8, color=PALETTE['gray'])
plt.tight_layout()
save_figure(fig, 'fig_boxplot_nivel_discapacidad.png')
plt.show()
```

## 4. Heatmap — disponibilidad y valores por país × categoría principal

```python
pivot = df[df['sexo']=='Total'][df['grupo_edad']=='Total'].pivot_table(
    index='pais_cod', columns='dis_nivel', values='pct_uso_internet'
)
fig, ax = plt.subplots(figsize=(10, 6))
sns.heatmap(pivot, annot=True, fmt='.1f', cmap='RdYlGn', ax=ax, linewidths=0.5,
            cbar_kws={'label': '% uso Internet'})
ax.set_title('Uso de Internet por país y nivel de discapacidad (2024)', fontweight='bold', color=PALETTE['navy'])
fig.text(0.01, -0.02, 'Fuente: elaboración propia a partir de Eurostat DSB_ICTIU01 (2024).', fontsize=8)
plt.tight_layout()
save_figure(fig, 'fig_heatmap_pais_discapacidad.png')
plt.show()
```
